```python
# Create an Object type Encoder to encode the object type columns present in the dataset
from sklearn.preprocessing import LabelEncoder
encodeObject = LabelEncoder()
countEncodes = 0
for col in dataset.columns[1:]:
    if dataset[col].dtype == 'object':
        encodeObject.fit(dataset[col])
        dataset[col] = encodeObject.transform(dataset[col])
        countEncodes += 1
```

```python
dataset
```

```python
# Create a Boolean type Encoder to encode the bool type columns present in the dataset
encodeBool = LabelEncoder()
countEncodes = 0
for col in dataset.columns[1:]:
    if dataset[col].dtype == 'bool':
        encodeBool.fit(dataset[col])
        dataset[col] = encodeBool.transform(dataset[col])
        countEncodes += 1
```

To be easier to use the dataset we decided to encode all the boolean values to have a better uniformity in the data. Previously we encoded all the object values since there were many date type columns. Using the encoders we have managed to achieve a better and more uniform dataset.

```python
#Fill all the nan values with the median value in each column
dataset = dataset.fillna(dataset.median())
```

```python
#Pass all float64 to int64
for col in dataset.columns[1:]:
    if dataset[col].dtype == 'float64':
        dataset[col] = dataset[col].astype('int64')
```

```python
#Count nan values in all columns
for i in dataset.columns:
    print(i,dataset[i].isnull().sum())
```

With this step we can now check that we have all the data encoded and that there are no missing values after the use of the KNN imputer to complete the dataset
